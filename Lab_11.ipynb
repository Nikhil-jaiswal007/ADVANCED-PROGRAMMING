# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
import itertools
def tsp(graph, start=0):
    n = len(graph)

    dp = {}

    for i in range(n):
        if i == start:
            continue
        dp[(1 << i, i)] = (graph[start][i], start)
    for r in range(2, n):
        for subset in itertools.combinations([i for i in range(n) if i != start], r):
            mask = sum([1 << i for i in subset])
            for j in subset:
                prev_mask = mask ^ (1 << j)
                candidates = []
                for k in subset:
                    if k == j:
                        continue
                    if (prev_mask, k) in dp:
                        candidates.append((dp[(prev_mask, k)][0] + graph[k][j], k))
                dp[(mask, j)] = min(candidates)

    mask = (1 << n) - 1 - (1 << start)
    candidates = []
    for j in range(n):
        if j == start:
            continue
        candidates.append((dp[(mask, j)][0] + graph[j][start], j))
    opt_cost, last_city = min(candidates)
    path = [start]
    mask = (1 << n) - 1 - (1 << start)
    while mask:
        path.append(last_city)
        new_mask = mask ^ (1 << last_city)
        _, prev_city = dp[(mask, last_city)]
        last_city = prev_city
        mask = new_mask
    path.append(start)
    path.reverse()

    return opt_cost, path
n = int(input("Enter number of cities: "))
print("Enter the distance matrix row by row:")
graph = []
for _ in range(n):
    row = list(map(int, input().split()))
    graph.append(row)

cost, path = tsp(graph, start=0)
print("Minimum cost:", cost)
print("Path:", path)
```
